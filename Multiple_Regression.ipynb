# Multiple Regression

```python
#Load your packages
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pa
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm


%matplotlib inline
```

# Time For Data 

```python
data = pa.read_csv('bikes.csv') #this is the file from exam 1
```

```python
data
```

```python
#Let's try to predict the time of the ride based on the speeds
#I need the data in numpy arrays rather than panda data frames.  This is what I do here
X = np.array(data.iloc[:,5:7])
y = np.array(data.iloc[:,4])
```

```python
X #no more leading indexes...
```

```python
#sklearn is a really amazing package! But is a little tricky to code.  You should first define the regressor
linreg = LinearRegression()#define it as a regular linear regression
linreg.fit(X,y)#This fits it!
linreg.score(X,y)#This is the R^2 value
```

```python
#You can make predictions
somevalues = [[13,35]]
linreg.predict(somevalues)
```

```python
#you can get the coefficients
linreg.coef_
```

```python
#and the intercept
linreg.intercept_
```

```python
#I can now do crazier features
X = np.append(X,X**2,axis = 1) #Here I square each value and gle that onto the end of X
```

```python
X = np.insert(X,-1,[X[:,0]*X[:,1]],axis = 1) #Here i multiply the first two values and glue that ontu the end of X.
```

```python
X
```

```python
linreg.fit(X,y)#Linear Regress again!
linreg.score(X,y)
```

```python
linreg.coef_
```

```python
linreg.intercept_
```

# How About Another Package?

```python
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```

```python
residuals = 0

for i in range (0,len(y)):
    residuals = residuals + (y[i]-predictions[i])**2
```

```python
residuals
```

Woof!  They do not agree!  Does not give an intercept though!  So add a column of 1s to the array.

```python
X1 = np.insert(X,0,np.ones(len(X[:,0])),axis = 1)
```

```python
model = sm.OLS(y, X1).fit()
predictions = model.predict(X1) # make the predictions by the model

# Print out the statistics
model.summary()
```
