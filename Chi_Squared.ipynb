```python
import numpy as np
from scipy import stats
import matplotlib.pylab as py

%matplotlib inline
```

```python
stats.randint.rvs(1,7)
```

```python
dist = []

for i in range (0,10000):
    dist.append(np.mean(stats.randint.rvs(1,7,size = 10)))
```

```python
dist
```

```python
mean = 3.5
sigma = np.sqrt(17.5/6)/np.sqrt(10)
```

```python
x = np.arange(0,7,0.001)
py.hist(dist,bins=10)
py.plot(x,1000*stats.norm.pdf(x,mean,sigma))
```

```python
sample =[]
for x in range(0,50):
    p = stats.randint.rvs(0,len(dist))
    sample.append(dist[p])
```

```python
sample
```

```python
newdist = []

for j in range (0,1000):
    sample =[]
    for x in range(0,20):
        p = stats.randint.rvs(0,len(dist))
        sample.append(dist[p])
    newdist.append(19*np.std(sample,ddof=1)**2/sigma**2)
```

```python
#Sturges Rule suggests that the number of bins should be
k = 1 + 3.322*np.log10(1000)
```

```python
x = np.arange(0,80,0.001)
py.hist(newdist,bins=35)
py.plot(x,1000*stats.chi2.pdf(x,19))
```
