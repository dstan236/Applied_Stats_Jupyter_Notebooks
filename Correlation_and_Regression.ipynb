# Time For Data 

```python
data = pa.read_csv('bikes.csv') #this is the file from exam 1
```

```python
data
```

```python
#Let's use something that is perfectly correlated!  minutes and seconds.
plt.scatter(data["Minutes"],data["Seconds"])
```

```python
#Correlation will be the pearsonr function it also returns your p value
stats.pearsonr(data["Minutes"],data["Seconds"])
```

```python
#Let's check that the p value is the same as I showed in class\
r = stats.pearsonr(data["Minutes"],data["Seconds"])[0]
n = len(data["Minutes"])
t = r/np.sqrt((1-r**2)/(n-2))
1-stats.t.cdf(t, n-2)
```

```python
#That one was boring, let's try again. This time with average speed and top speed
plt.scatter(data["AvgSpeed"],data["TopSpeed"])
```

```python
#Woof!
stats.pearsonr(data["AvgSpeed"],data["TopSpeed"])
```

```python
#Still showing very low p value!
#Let's check that the p value is the same as I showed in class\
r = stats.pearsonr(data["AvgSpeed"],data["TopSpeed"])[0]
n = len(data["Minutes"])
t = r/np.sqrt((1-r**2)/(n-2))
1-stats.t.cdf(t, n-2)
```

```python
#Not the same?  It is two tailed!
2*(1-stats.t.cdf(t, n-2))
```

# Time for Regression Lines

```python
stats.linregress(data["AvgSpeed"],data["TopSpeed"])
```

```python
#Easy right?  Let's make it into something useable too
m = stats.linregress(data["AvgSpeed"],data["TopSpeed"])[0]
b = stats.linregress(data["AvgSpeed"],data["TopSpeed"])[1]
def y(x):
    return(m*x+b)
    
```

```python
y(3)
```

```python
x = np.arange( 13,17,.01)
plt.scatter(data["AvgSpeed"],data["TopSpeed"])
plt.plot(x,y(x))
```

```python
#sum of the residuals
s = 0
for i in range (0,len(data["AvgSpeed"])):
    s = s+y(data["AvgSpeed"][i])-data["TopSpeed"][i]
```

```python
#Sum of Squares of Residuals
ss = 0
for i in range (0,len(data["AvgSpeed"])):
    ss = ss+(y(data["AvgSpeed"][i])-data["TopSpeed"][i])**2
```
